```python
# Imports and setup for image loading
import h5py
import numpy as np

H5_PATH = "Galaxy10_DECals(1).h5"

# Load Data
with h5py.File(H5_PATH, "r") as f:
    images = np.array(f["images"])
    labels = np.array(f["ans"])
```

```python
import matplotlib.pyplot as plt
class_names = ['Disturbed', 
               'Merging', 
               'Round Smooth', 
               'In-between Round Smooth', 
               'Cigar Round Smooth', 
               'Barred Spiral', 
               'Unbarred Tight Spiral', 
               'Unbarred Loose Spiral', 
               'Edge-on without Bulge', 
               'Edge-on with Bulge'
              ]

def plot_classes(images, labels, class_names):
    fig, axes = plt.subplots(2, 5, figsize=(20, 8))
    axes = axes.flatten()
    
    for i in range(len(class_names)):
        idx = np.where(labels == i)[0][0]
        img = images[idx]
        
        axes[i].imshow(img)
        axes[i].set_title(f"Class {i}:\n{class_names[i]}", fontsize=10)
        axes[i].axis('off')
        
    plt.tight_layout()
    plt.show()

plot_classes(images, labels, class_names)
```

```python
import seaborn as sns

def plot_class_distribution(labels, class_names):
    plt.figure(figsize=(12, 6))
    counts = np.bincount(labels)
    
    sns.barplot(x=class_names, y=counts, palette="viridis")
    plt.xticks(rotation=45, ha='right')
    plt.xlabel("Galaxy Classification")
    plt.ylabel("Number of Images")
    plt.title("Galaxy10 DECals Class Distribution")
    
    for i, count in enumerate(counts):
        plt.text(i, count + 20, str(count), ha='center', fontweight='bold')
        
    plt.tight_layout()
    plt.show()

plot_class_distribution(labels, class_names)
```
